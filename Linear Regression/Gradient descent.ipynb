```python
import numpy as np
import time
import matplotlib.pyplot as plt

def gradient_descent(learning_rate, iterations):
# List số lần lặp
  a = []
# List giá trị của f(x) = x*x
  b = []
  x0 = 2
  for i in range(iterations):
      y = x0*x0
      a.append(i)
      b.append(y) 
      x0 -= learning_rate*(2*x0)
  
  plt.title('Learning rate:'+ str(learning_rate),c='red')
  plt.xlabel('Iterations')
  plt.ylabel('Value')
  plt.plot(a,b)
  plt.show()

gradient_descent(0.001,1000)
gradient_descent(0.01,1000)
gradient_descent(0.1,1000)


```
