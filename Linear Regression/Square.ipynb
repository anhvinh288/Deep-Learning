```python
import numpy as np
import time
import matplotlib.pyplot as plt
import pandas as pd

# Import data
data_path = "drive/My Drive/Colab/Linear regression/data_square.csv"
data = pd.read_csv(data_path)

# Lấy số lượng dataset
N = data.shape[0]

# Lấy cột diện tích
x = data.iloc[:,0]
x = x.values.reshape(-1,1)
xg = x
# Lấy cột giá nhà
y = data.iloc[:,1]
y = y.values.reshape(-1,1)

# Vẽ đồ thị
plt.scatter(x, y, c='b')
plt.xlabel('Diện tích')
plt.ylabel('Giá nhà')


# Thêm cột 1 vào x [[1 x x^2],[1 x x^2],...]
# z = np.ones((N,1))
# x1 = np.insert(x,[0],z,axis=1)
x = np.hstack((np.ones((N, 1)),x, x*x))

# Khởi tạo parameters, learning_rate và iteration
w = np.array([1.,1.,1.]).reshape(-1,1)
learning_rate = 0.0000000002
iteration = 20000000
# 10000000
# Tạo hàm Cost function
cost = np.zeros((iteration,1))
num = [0]
# Áp dụng gradient descent
for i in range(1,iteration):
        s = np.dot(x,w) - y
        cost[i] = 0.5*np.sum(s*s)
        # w[0] -= learning_rate*np.sum(s)
        # w[1] -= learning_rate*np.sum(np.multiply(s,x[:,1].reshape(-1,1)))
        # w[2] -= learning_rate*np.sum(np.multiply(s,x[:,2].reshape(-1,1)))
        w -= learning_rate*(np.matmul(np.transpose(x),s))
        num.append(i)

print(w)
predict = np.dot(x,w)

plt.plot(xg[:],predict[:], 'r')
plt.show() 
# Vẽ đồ thị gradient descent
plt.plot(num,cost,c='red')
plt.xlabel("Iteration")
plt.ylabel("Cost function")
plt.plot()
print(cost[iteration-1])

```
